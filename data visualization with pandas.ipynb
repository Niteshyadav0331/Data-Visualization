## Data Visualization With Pandas

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
%matplotlib inline
```

#### Data Generation

```python
from numpy.random import randn, randint, uniform, sample
```

```python
df = pd.DataFrame(randn(1000), index = pd.date_range('2019-06-07', periods = 1000), columns = ['value'])
```

```python
df.head(8)
```

```python
ts = pd.Series(randn(1000), index = pd.date_range('2019-06-07', periods = 1000))
ts.head(8)
```

### Line Plot

```python
df.value = df.value.cumsum()
df.head()
```

```python
df.plot()
```

```python
df.plot(figsize = (10, 5))
```

```python
ts = ts.cumsum()
ts.head()
```

```python
ts.plot()
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris.plot(figsize = (10, 5))
```

```python
x = iris.drop(labels = ['sepal_width', 'petal_width'], axis = 1)
x.head()
```

```python
y = iris[['sepal_width', 'petal_width']]
y.head()
```

```python
ax = x.plot()
y.plot(figsize = (16, 6), secondary_y = True, ax = ax)
```

### Bar Plot

```python
df = iris.drop(['species'], axis = 1)
df.head()
```

```python
df.iloc[0].plot(kind = 'bar')
```

```python
df.iloc[0].plot.bar()
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['pclass'].plot.hist()
```

```python
titanic['survived'].plot(kind = 'hist')
```

### Stacked Plot

```python
df = pd.DataFrame(randn(10, 4), columns = ['a', 'b', 'c', 'd'])
df.head()
```

```python
df.plot.bar(stacked = True,  figsize = (12, 6))
```

```python
df.plot.barh(stacked = True, figsize =(12, 6))
```

### Histogram

```python
iris.plot(kind = 'hist', alpha = 0.8)
```

```python
iris.plot(kind = 'hist', bins = 50, alpha = 0.7)
```

```python
iris.plot(kind = 'hist', bins = 50, stacked = True, alpha = 0.7)
```

```python
df = iris.drop(['species'], axis = 1)
df.head()
```

```python
df.hist()
```

### Box Plot

```python
df.plot(kind = 'box', figsize = (10, 5))
```

```python
df.plot(kind = 'box', figsize = (10, 5), vert = False)
```

### Area and Scatter Plot

```python
df.head()
```

```python
df.plot(kind = 'area', figsize = (10, 6))
```

```python
df.plot(kind = 'area', stacked = False, alpha = 0.7, figsize = (10, 6))
```

```python
df.plot.scatter(x = 'sepal_width', y = 'petal_width')
```

```python
df.plot.scatter(x = 'sepal_width', y = 'petal_width', c = 'sepal_length')
```

```python
ax = df.plot.scatter(x = 'sepal_width', y = 'petal_width', label = 'width')
df.plot.scatter(x = 'sepal_length', y = 'sepal_width', ax = ax, color = 'r', figsize = (10, 5))
```

### Hex and Pie Plot

```python
df.plot.hexbin(x = 'sepal_length', y = 'petal_length', gridsize = 10, figsize = (10, 6), C = 'sepal_width')
```

```python
df.iloc[0]
```

### Scatter Matrix and Subplots

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df, figsize = (10,10), diagonal = 'kde', color = 'r')
plt.show()
```

```python
df.plot()
```

```python
df.plot(subplots = True)
plt.show()
```

```python
df.plot(subplots = True, sharex = False)
plt.tight_layout()
plt.show()
```

```python
df.plot(subplots = True, sharex = False, layout = (2, 2), figsize = (10, 5))
plt.tight_layout()
plt.show()
```
