# `Time Series` Plot 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = [8, 4]
plt.rcParams['figure.dpi'] = 140
```

```python
df = pd.read_csv('daily-temperature.csv')
df
```

```python
df.info()
```

**There are no null values in the dataset and datatype of Date column is object so we have to convert it to datetime.**

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.info()
```

```python
df.set_index('Date', inplace = True)
df.head()
```

### Line and Scatter Plot

```python
df.plot(color = 'r')
plt.ylabel('Temp')
plt.title('Line Plot')
plt.show()
```

```python
df.plot(style = '.', color = 'red')
plt.ylabel('Temp')
plt.title('Scatter Plot')
plt.show()
```

### Subplots

```python
groups = df.groupby(pd.Grouper(freq = 'A'))
```

```python
keys = groups.groups.keys()
keys
```

```python
groups.get_group('1982-12-31')
```

```python
for key in keys:
    print(key)
```

```python
key
```

```python
groups.get_group(key)['Temp'].values
```

```python
years = pd.DataFrame()
for key in keys:
    years[key] = groups.get_group(key)['Temp'].values
    
years
```

```python
years.plot(subplots = True, sharex = False, figsize = (10, 30))
plt.tight_layout()
plt.show()
```

### Heatmap

```python
years
```

```python
plt.matshow(years.T, aspect = 'auto')
plt.show()
```

**From this heatmap we can say that at beginning and ending of year we have hot weather and in mid we have cold weather.**
