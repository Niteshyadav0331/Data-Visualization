# Plotly and Cufflinks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from plotly.offline import iplot
import plotly as py
```

```python
import cufflinks as cf
```

```python
py.offline.init_notebook_mode(connected = True)
```

```python
cf.go_offline()
```

### Line Plot

```python
df = pd.DataFrame(np.random.randn(100, 3), columns = ['a', 'b', 'c'])
df.head()
```

```python
df = df.cumsum()
df.head()
```

```python
df.iplot()
```

### Scatter Plot

```python
df.head()
```

```python
df.iplot(x = 'a', y = 'b', mode = 'markers', size = 10, title = 'Scatter Plot', xTitle = 'X axis', yTitle = 'Y axis')
```

### Bar Plot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.iplot(x = 'sex', y = 'survived', kind = 'bar', title = 'Bar Plot')
```

```python
titanic.sex.value_counts()
```

```python
df.iplot(kind = 'bar')
```

```python
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
df.iplot(kind = 'barh', barmode = 'stack', bargap = 0.5)
```

### Box Plot and Area Plot

```python
df.iplot(kind = 'box')
```

```python
df.iplot(kind = 'area')
```

```python
df.iplot(kind = 'area', fill = True)
```

### 3D Plot

```python
df.iplot(kind = 'surface')
```

```python
df.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind = 'surface')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind = 'surface', colorscale = 'rdylbu')
```

### Spread Plot and Hist Plot 

```python
df.iplot(kind = 'spread')
```

```python
df.iplot(kind = 'hist', bins = 25, bargap = 0.5)
```

```python
df.iplot(kind = 'hist', barmode = 'stack', bargap = 0.5, bins = 25)
```

```python
df.iplot(kind = 'hist', bins = 25, barmode = 'group', bargap = 0.5)
```

### Bubble Plot and Heatmap

```python
bb = cf.datagen.bubble3d(5, 4, mode = 'stocks')
```

```python
bb.head()
```

```python
bb.iplot(kind = 'bubble3d', x = 'x', y = 'y', z = 'z', size = 'size')
```

```python
df.corr()
```

```python
df.corr().iplot(kind = 'heatmap', colorscale = 'spectral')
```

```python
cf.datagen.heatmap(20, 20).iplot(kind = 'heatmap', colorscale = 'spectral')
```
